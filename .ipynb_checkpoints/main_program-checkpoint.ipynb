```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing custome classes created by me to scale,to encode,to split data into training and testing sets,to select thoes features that are playing important part in output
from preprocessing import splitter,Encoder,scaler
from feature_selection import selector
from metrics import matrix
sp=splitter()
en=Encoder()
sc=splitter()
slt=selector()
mt=matrix()
```

```python
#importing breast cancer data from usi repositary
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data"
names = ['id', 'clump_thickness', 'uniform_cell_size', 'uniform_cell_shape',
       'marginal_adhesion', 'single_epithelial_size', 'bare_nuclei',
       'bland_chromatin', 'normal_nucleoli', 'mitoses', 'class']
dataset = pd.read_csv(url, names=names)
dataset.shape
```

```python
#in dataset some attibutes are '?' so to encode that data we need to perform this 
for column in dataset.columns:
    if dataset[column].dtype == type(object):
        dataset[column] = en.labelEncoding(dataset[column])
              
```

```python
#deviding dataset into dependant and independant dataset
x=dataset.iloc[:,1:10].values
y=dataset.iloc[:,10].values
```

```python
#visualising 1st attribute of independant data with dependant
plt.scatter(x[:,1],y,color='red')
plt.show()
```

```python
#visualizing 2nd attribute with dependant data
plt.scatter(x[:,2],y,color='red')
plt.show()
```

```python
#to visualize each attribute of dataset
dataset.hist(figsize=(20,20))
plt.show()
```

```python
#applying feature selection
n=int(input("enter a no. of colomns you want to take in a model"))
best=slt.select(x,y,n)
```

```python
#printing the score of each attribute according to importance in output
np.set_printoptions(precision=3)
print(best.scores_)
```

```python
#making our final matrix of features
FinalX=best.transform(x)
FinalX
```

```python
#splitting dataset into training and testing sets
x_train,x_test,y_train,y_test=sp.decomposition(FinalX,y,test_size=0.10,random_state=0)
```

```python
#creating our SVM classification model
from sklearn.svm import SVC
cls=SVC(kernel='linear',random_state=0)
cls.fit(x_train,y_train)
```

```python
#predicting values for x_test
y_pred=cls.predict(x_test)
```

```python
#cheaking accuracy of model
accuracy=mt.accuracy(y_pred,y_test)
accuracy
```
